# Matplotlib - Plots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,20,21)
y1 = 2*x**2
y2 = x**2 - 3*x

plt.figure(figsize=(5,3),dpi=100)
plt.plot(x,y1,lw=1,color='red',linestyle='-',label='y1')
plt.plot(x,y2,lw=1,color='green',linestyle='-',label='y2')
plt.legend(loc='upper center')
plt.title("y=f(x)",fontsize=10)
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(x[0],x[-1])
plt.grid()
plt.show()
```

# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0,360,30)
y1 = np.sin(theta*np.pi/180)
y2 = np.cos(theta*np.pi/180)

x = np.linspace(0,10,11)
z1 = x**3
z2 = x**2

plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(theta,y1)
plt.title('cos'r'($\theta{}$)')
plt.tight_layout(pad=4.0)
plt.subplot(222)
plt.plot(theta,y2)
plt.title('sin'r'($\theta{}$)')

plt.subplot(223)
plt.plot(x,z1)
plt.title('Z = 'r'$x^{3}$')
plt.tight_layout(pad=4)
plt.subplot(224)
plt.plot(x,z2)
plt.title('Z = 'r'$x^{2}$')

plt.show()
```

# Conntour Plots

```python
import numpy as np
import matplotlib.pyplot as plt

xp = np.linspace(-5,5,5)
yp = np.linspace(-5,5,5)
X,Y = np.meshgrid(xp,yp)
plt.scatter(X,Y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

xp = np.linspace(-5,5,5)
yp = np.linspace(-5,5,5)
X,Y = np.meshgrid(xp,yp)
Z = np.sin(X) + np.cos(Y)
plt.contour(X,Y,Z,10,cmap='jet')
plt.colorbar()
plt.show()
```

# Surface Plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)
Z = (X**2+Y**2)

plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
mysurf = ax.plot_surface(X,Y,Z,cmap='jet',alpha=0.8)
plt.colorbar(mysurf)
cset = plt.contourf(X,Y,Z,zdir='z',offset=5,cmap='jet')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```
